```python
train_set['Journey_Date'] = pd.to_datetime(train_set['Date_of_Journey'], format ='%d/%m/%Y').dt.day

train_set['Journey_Month'] = pd.to_datetime(train_set['Date_of_Journey'], format ='%d/%m/%Y').dt.month

#train_set['Journey_Year'] = pd.to_datetime(train_set['Date_of_Journey'], format ='%d/%m/%Y').dt.year
```

```python
train_set = train_set.drop(columns = ['Date_of_Journey'])
```

```python
train_set['Dep_Hour'] = pd.to_datetime(train_set['Dep_Time'],format= '%H:%M').dt.hour

train_set['Dep_Min'] = pd.to_datetime(train_set['Dep_Time'],format= '%H:%M').dt.minute

train_set= train_set.drop(columns = ['Dep_Time'])
```

```python
train_set['Arrival_Hour'] = pd.to_datetime(train_set['Arrival_Time']).dt.hour

train_set['Arrival_Min'] = pd.to_datetime(train_set['Arrival_Time']).dt.minute

train_set= train_set.drop(columns = ['Arrival_Time'])
```

```python
train_set
```

```python
test_set['Journey_Date'] = pd.to_datetime(test_set['Date_of_Journey'], format= '%d/%m/%Y').dt.day
test_set['Journey_Month'] = pd.to_datetime(test_set['Date_of_Journey'], format= '%d/%m/%Y').dt.month
#test_set['Journey_Year'] = pd.to_datetime(test_set['Date_of_Journey'], format= '%d/%m/%Y').dt.year

test_set = test_set.drop(columns = ['Date_of_Journey'])
```

```python
test_set['Dep_Hour'] = pd.to_datetime(test_set['Dep_Time'],format= '%H:%M').dt.hour

test_set['Dep_Min'] = pd.to_datetime(test_set['Dep_Time'],format= '%H:%M').dt.minute

test_set = test_set.drop(columns= ['Dep_Time'])
```

```python
test_set['Arrival_Hour'] = pd.to_datetime(test_set['Arrival_Time']).dt.hour

test_set['Arrival_Min'] = pd.to_datetime(test_set['Arrival_Time']).dt.minute

test_set= test_set.drop(columns = ['Arrival_Time'])
```

```python
test_set
```

```python
X_train = train_set.drop(columns=['Price', 'Duration'])
```

```python
X_train.info()
```

```python
X_test = test_set.drop(columns=['Duration'])
```

```python
X_test.info()
```

```python
y_train = train_set['Price']
y_train.head()
```

```python
# [0] means we are just taking the labels, eg 0,2,3,1
X_train['Airline'] = pd.factorize(X_train['Airline'])[0] 
X_train['Source']= pd.factorize(X_train['Source'])[0]
X_train['Destination']= pd.factorize(X_train['Destination'])[0]
X_train['Route']= pd.factorize(X_train['Route'])[0]
X_train['Total_Stops']= pd.factorize(X_train['Total_Stops'])[0]
X_train['Additional_Info']= pd.factorize(X_train['Additional_Info'])[0]
X_train.head()
```

```python
X_test['Airline'] = pd.factorize(X_test['Airline'])[0] 
X_test['Source']= pd.factorize(X_test['Source'])[0]
X_test['Destination']= pd.factorize(X_test['Destination'])[0]
X_test['Route']= pd.factorize(X_test['Route'])[0]
X_test['Total_Stops']= pd.factorize(X_test['Total_Stops'])[0]
X_test['Additional_Info']= pd.factorize(X_test['Additional_Info'])[0]
X_test.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()

X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
import statsmodels.api as sm
model = sm.OLS(y_train, X_train.astype(int))
model = model.fit()
print(model.summary())
```

```python
print(X_train.shape, X_test.shape, y_train.shape)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
print('Parameters: ', model.params)
print('R2: ', model.rsquared)
```

```python
pd.DataFrame(y_pred, columns = ['Price']).to_excel("Predicted_price.xlsx", index = False)
```
