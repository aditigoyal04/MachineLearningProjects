```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('HR-Employee-Attrition.csv')
dataset.head()
```

```python
dataset.isnull().sum()
dataset.dtypes
dataset.info()
```

```python
%matplotlib inline
pd.plotting.scatter_matrix(dataset)
```

```python
# Converting all the categorical to numerical data
Num_val = {'Yes':1, 'No':0}
dataset['Attrition'] = dataset["Attrition"].apply(lambda x: Num_val[x])
Dataset_categorical = dataset[['Attrition','BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','Over18','OverTime']]
Dataset_categorical = pd.get_dummies(Dataset_categorical)
Dataset_categorical.head()
```

```python
Dataset_numerical = dataset[['Age','DailyRate','Education','EmployeeCount','EmployeeNumber', 'HourlyRate', 'JobLevel',
                            'MonthlyIncome','MonthlyRate','NumCompaniesWorked','PercentSalaryHike','PerformanceRating',
                            'StandardHours','TotalWorkingYears']]
Dataset_numerical.head()
```

```python
New_dataset = pd.concat([Dataset_categorical,Dataset_numerical],axis = 1)
New_dataset.head()
```

```python
y = New_dataset['Attrition']
X = New_dataset.drop('Attrition', axis = 1)
y.head()
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.20,random_state= 42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
from sklearn.metrics import accuracy_score
ac = accuracy_score(y_pred,y_test)
ac
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier_r = RandomForestClassifier()
classifier_r.fit(X_train,y_train)
y_pred_r = classifier.predict(X_test)

print(accuracy_score(y_pred_r, y_test))
```

```python
graph = pd.Series(classifier_r.feature_importances_, index=X.columns)
graph = graph.nlargest()
graph.plot(kind='barh')
```
