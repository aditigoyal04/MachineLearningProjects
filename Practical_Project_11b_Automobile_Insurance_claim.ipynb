```python
df = df.drop(columns = ['Effective To Date'])
```

```python
df.head()
```

```python
df.nlargest(5, ['Total Claim Amount'])
```

```python
df['Claim Amount'] = df['Claim Amount'].astype(np.int)
df['Total Claim Amount'] = df['Total Claim Amount'].astype(int)
```

```python
#X = df.drop(columns = ['Customer','State','Effective year','Education','State Code','Marital Status','Country','Income','Claim Amount','EmploymentStatus','Gender'])
X = df.drop(columns = ['Customer','State Code','Effective year','Country'])
y = df['Total Claim Amount']
```

```python
X.head(2)
```

```python
#y.head()
```

```python
#X['State Code'] = pd.factorize(X['State Code'])[0]
#X['Country'] = pd.factorize(X['Country'])[0]
X['State'] = pd.factorize(X['State'])[0]
X['Response'] = pd.factorize(X['Response'])[0]
X['Coverage'] = pd.factorize(X['Coverage'])[0]
X['Education'] = pd.factorize(X['Education'])[0]
X['EmploymentStatus'] = pd.factorize(X['EmploymentStatus'])[0]
X['Gender'] = pd.factorize(X['Gender'])[0]
X['Location Code'] = pd.factorize(X['Location Code'])[0]
X['Marital Status'] = pd.factorize(X['Marital Status'])[0]
X['Policy Type'] = pd.factorize(X['Policy Type'])[0]
X['Policy'] = pd.factorize(X['Policy'])[0]
X['Claim Reason'] = pd.factorize(X['Claim Reason'])[0]
X['Sales Channel'] = pd.factorize(X['Sales Channel'])[0]
X['Vehicle Class'] = pd.factorize(X['Vehicle Class'])[0]
X['Vehicle Size'] = pd.factorize(X['Vehicle Size'])[0]
```

```python
X.head(2)
```

```python
y.head(2)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state= 0)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test= scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)
reg.score(X_test,y_test) #Overfitting of dataset
```

```python
print(reg.intercept_)
```

```python
print(reg.coef_)
```

```python
import statsmodels.api as sm
model = sm.OLS(y_train,X_train).fit()
print(model.summary())
```

```python
from sklearn.svm import SVR
sv = SVR()
sv.fit(X_train,y_train)
```

```python
y_p = sv.predict(X_test)
sv.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dc = DecisionTreeRegressor()
dc.fit(X_train,y_train)
```

```python
y_d = dc.predict(X_test)
dc.score(X_test,y_test) #Overfitting of dataset
```

```python
from sklearn.ensemble import RandomForestRegressor
random = RandomForestRegressor()
random.fit(X_train,y_train)

#random.score(X_test,y_test) # overfitting
```
