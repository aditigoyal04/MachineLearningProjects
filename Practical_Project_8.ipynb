```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, accuracy_score
```

```python
dataset = pd.read_csv('data_banknote_authentication.txt', header = None)
dataset.head()
```

```python
dataset.info()
dataset.describe()
dataset[4].value_counts()
```

```python
dataset.sample(n=10)
```

```python
# a random 1005 of the sample(frac = 1)
dataset = dataset.sample(frac=1)
dataset
```

```python
X = dataset.iloc[:,0:-1].values
y = dataset.iloc[:,-1].values
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30, random_state=0)
```

```python
print(X_test.shape, X_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train,y_train)

y_pred = classifier.predict(X_test)
ac = classifier.score(X_test,y_test)
print(ac)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier_r = RandomForestClassifier()
classifier_r.fit(X_train,y_train)

y_pred_r = classifier_r.predict(X_test)
ac = classifier_r.score(X_test,y_test)
print(ac) 
```

```python
from sklearn.tree import DecisionTreeClassifier
deci = DecisionTreeClassifier(criterion= 'gini')
deci.fit(X_train,y_train)

d_y_pred = deci.predict(X_test)
print(deci.score(X_test,y_test))
```
