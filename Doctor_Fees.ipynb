```python
train_set['Year_of_Exp'] = train_set['Experience'].str.extract(r'^(\d+)', expand=False)
train_set = train_set.drop(columns = ['Experience'])

train_set['Rating'] = train_set['Rating'].str.extract(r'^(\d+)', expand=False)
train_set['Rating'].fillna('-1', inplace= True)

train_set['Place'].fillna('Unknown', inplace= True)
train_set['Place']=train_set['Place'].str.split(", ") 
  
train_set['Location']=train_set['Place'].str[0] 
train_set['City']=train_set['Place'].str[-1]
train_set.loc[train_set['City'] == 'e']
train_set['City'][3980]= 'Unknown'
train_set = train_set.drop(columns = ['Place'])
```

```python
print(train_set['Rating'].unique())

Rate = []
for i in train_set['Rating']:
    if i>='90':
        Rate.append('95')
    elif i>='80' and i<'90':
        Rate.append('85')
    elif i>='70' and i<'80':
        Rate.append('75')
    elif i>='60' and i<'70':
        Rate.append('65')
    elif i>='50' and i<'60':
        Rate.append('55')
    elif i>='40' and i<'50':
        Rate.append('45')
    elif i>='30' and i<'40':
        Rate.append('35')
    elif i>='20' and i<'30':
        Rate.append('25')
    elif i>='10' and i<'20':
        Rate.append('15')
    elif i>='0' and i<'10':
        Rate.append('5')
    else:
        Rate.append('-1')
train_set['Rating']=Rate
```

```python
train_set['Qualification']=train_set['Qualification'].str.split(", ") 
  
train_set['Quali_1']=train_set['Qualification'].str[0] 
train_set['Quali_2']=train_set['Qualification'].str[1] 
train_set['Quali_3']=train_set['Qualification'].str[2] 

train_set['Quali_2'].fillna('Unavailable', inplace= True)
train_set['Quali_3'].fillna('Unavailable', inplace= True)

train_set=train_set.drop(columns = ['Qualification'])
```

```python
test_set['Years_of_Experience'] = test_set['Experience'].str.extract(r'^(\d+)', expand=False)
test_set = test_set.drop(columns = ['Experience'])

test_set['Rating'] = test_set['Rating'].str.extract(r'^(\d+)', expand=False)
test_set['Rating'].fillna('-1', inplace= True)

test_set['Place'].fillna('Unknown', inplace= True) 
test_set['Place']= test_set['Place'].str.split(", ", 1)
  
test_set['Location']=test_set['Place'].str.get(0) 
test_set['City']=test_set['Place'] .str.get(1) 
test_set['City'].fillna('Unknown', inplace = True)
test_set = test_set.drop(columns = ['Place'])
```

```python
print(test_set['Rating'].unique())

Rate = []
for i in test_set['Rating']:
    if i>='90':
        Rate.append('95')
    elif i>='80' and i<'90':
        Rate.append('85')
    elif i>='70' and i<'80':
        Rate.append('75')
    elif i>='60' and i<'70':
        Rate.append('65')
    elif i>='50' and i<'60':
        Rate.append('55')
    else:
        Rate.append('-1')
test_set['Rating']=Rate
```

```python
test_set['Qualification']=test_set['Qualification'].str.split(", ") 
  
test_set['Quali_1']=test_set['Qualification'].str[0] 
test_set['Quali_2']=test_set['Qualification'].str[1] 
test_set['Quali_3']=test_set['Qualification'].str[2] 

test_set['Quali_2'].fillna('Unavailable', inplace= True)
test_set['Quali_3'].fillna('Unavailable', inplace= True)

test_set=test_set.drop(columns = ['Qualification'])
```

```python
X_train = train_set.drop(columns = ['Fees','Miscellaneous_Info'])
y_train = train_set['Fees']

X_test = test_set.drop(columns = ['Miscellaneous_Info'])
```
