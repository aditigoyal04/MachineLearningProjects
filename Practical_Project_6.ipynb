```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/dsrscientist/DSData/master/SBI_Life_insurance.csv')
dataset.head()
#dataset.info() # it shows no missing values
#dataset.describe()
```

```python
dataset['sex'].unique()
dataset['children'].unique()
dataset['smoker'].unique()
dataset['region'].unique()
dataset['insuranceclaim'].unique()
```

```python
dataset['age'].value_counts()
sns.countplot(x='age',data = dataset)
plt.show()
```

```python
dataset['bmi'].value_counts()
sns.countplot(x='bmi',data = dataset)
plt.show()
```

```python
dataset['charges'].value_counts()
sns.countplot(x='charges',data = dataset)
plt.show()
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
print(X.shape, y.shape)
```

```python
dataset['age'].value_counts()
sns.countplot(x='age',data = dataset)
plt.show()

dataset['sex'].value_counts()
sns.countplot(x='sex',data = dataset)
plt.show()

dataset['bmi'].value_counts()
sns.countplot(x='bmi',data = dataset)
plt.show()

dataset['children'].value_counts()
sns.countplot(x='children',data = dataset)
plt.show()

dataset['smoker'].value_counts()
sns.countplot(x='smoker',data = dataset)
plt.show()

dataset['region'].value_counts()
sns.countplot(x='region',data = dataset)
plt.show()

dataset['charges'].value_counts()
sns.countplot(x='charges',data = dataset)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train, y_train)

y_pred = log.predict(X_test)
print('Accuracy Score :',log.score(X_test,y_test))
confusion_matrix(y_pred,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,y_train)

nb_y_pred = nb.predict(X_test)
print('Accuracy score :', nb.score(X_test,y_test))
confusion_matrix(nb_y_pred, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
deci = DecisionTreeClassifier(criterion= 'gini')
deci.fit(X_train,y_train)

d_y_pred = deci.predict(X_test)
print('Accuracy Score :', deci.score(X_test,y_test))
confusion_matrix(d_y_pred, y_test)
```

```python
from sklearn.svm import SVC
sv = SVC(kernel='linear')
sv.fit(X_train,y_train)

s_y_pred = sv.predict(X_test)
print('Accuracy Score :', sv.score(X_test,y_test))
confusion_matrix(s_y_pred, y_test)
```
